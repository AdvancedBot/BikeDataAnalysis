## Exercise 2(d) Use the simulation function

```python
import numpy as np
import timeit

def sim_uniform(n, n_reps, theta, rng):
    # Simulate random samples
    return rng.uniform(0, theta, size=(n_reps, n))

def est_uniform(x):
    # Calculate the MLE and the mean-based estimator
    mle = np.max(x, axis=1)
    mean_based = 2 * np.mean(x, axis=1)
    return np.column_stack((mle, mean_based))

def sim_est_uniform(n, n_reps, theta, rng):
    # Simulate random samples and estimate
    samples = sim_uniform(n, n_reps, theta, rng)
    estimates = est_uniform(samples)
    return estimates

def analyze_estimators(n, theta, n_reps, rng):
    # Simulate and estimate
    estimates = sim_est_uniform(n, n_reps, theta, rng)
    
    # Calculate correlation
    corr = np.corrcoef(estimates[:, 0], estimates[:, 1])[0, 1]
    
    # Calculate means
    mean_mle = np.mean(estimates[:, 0])
    mean_mean_based = np.mean(estimates[:, 1])
    
    # Calculate MSE
    mse_mle = np.mean((estimates[:, 0] - theta) ** 2)
    mse_mean_based = np.mean((estimates[:, 1] - theta) ** 2)
    
    # Calculate ratio of MSE
    mse_ratio = mse_mean_based / mse_mle

    return corr, mean_mle, mean_mean_based, mse_ratio

rng = np.random.default_rng(seed=1)
n_values = [10, 100, 1000]
theta = 3
n_reps = 10000
results = {}

for n in n_values:
    results[n] = analyze_estimators(n, theta, n_reps, rng)

# Output the results
for n, res in results.items():
    print(f"n={n}: Correlation={res[0]:.2f}, Mean MLE={res[1]:.2f}, Mean Mean-Based={res[2]:.2f}, MSE Ratio={res[3]:.2f}")
```

## Exercise 2(e) The modi_ed maximum likelihood estimator

```python
def est_uniform(x, n):
    # Calculate the maximum likelihood estimator
    mle = np.max(x, axis=1)
    # Calculate the mean-based estimator
    mean_based = 2 * np.mean(x, axis=1)
    # Calculate the unbiased estimator based on MLE
    unbiased_mle = mle * (n + 1) / n
    return np.column_stack((mle, mean_based, unbiased_mle))

def mse(estimates, theta):
    # Calculate the mean squared error for each column of estimates
    return np.mean((estimates - theta) ** 2, axis=0)

# You would then use these functions in your simulation:

n_values = [10, 100, 1000]
theta = 3
n_reps = 10000
rng = np.random.default_rng(seed=1)

for n in n_values:
    # Simulate samples
    samples = sim_uniform(n, n_reps, theta, rng)
    # Calculate estimators
    estimates = est_uniform(samples, n)
    # Compute MSE for each estimator
    mse_values = mse(estimates, theta)
    # Extract results
    mle_mse = mse_values[0]
    mean_based_mse = mse_values[1]
    unbiased_mle_mse = mse_values[2]
    
    # Output the results
    print(f"n={n}: MSE MLE={mle_mse:.2f}, MSE Mean-Based={mean_based_mse:.2f}, MSE Unbiased MLE={unbiased_mle_mse:.2f}")
```

## Exercise 4(f) Choosing a model with transformations

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Loading the dataset
bikeshare_data = pd.read_csv('Bikeshare.csv')

# Preparing the response variable
y = bikeshare_data['bikers']


# Model 1: Quadratic transformation of day
X1 = sm.add_constant(pd.DataFrame({
    'day': bikeshare_data['day'],
    'day_sq': bikeshare_data['day']**2
}))
model_1 = sm.OLS(y, X1).fit()

# Model 2: Cubic transformation of day
X2 = sm.add_constant(pd.DataFrame({
    'day': bikeshare_data['day'],
    'day_sq': bikeshare_data['day']**2,
    'day_cub': bikeshare_data['day']**3
}))
model_2 = sm.OLS(y, X2).fit()

# Model 3: Logarithmic transformation of day
X3 = sm.add_constant(pd.DataFrame({
    'log_day': np.log(bikeshare_data['day'])
}))
model_3 = sm.OLS(y, X3).fit()

# Model 4: Interaction term between day and temp
X4 = sm.add_constant(pd.DataFrame({
    'day': bikeshare_data['day'],
    'temp': bikeshare_data['temp'],
    'day_temp': bikeshare_data['day'] * bikeshare_data['temp']
}))
model_4 = sm.OLS(y, X4).fit()

# Model 5: Interaction term with quadratic transformation of day and temp
X5 = sm.add_constant(pd.DataFrame({
    'day_sq': bikeshare_data['day']**2,
    'temp': bikeshare_data['temp'],
    'day_sq_temp': (bikeshare_data['day']**2) * bikeshare_data['temp']
}))
model_5 = sm.OLS(y, X5).fit()

# Compiling the results
results = [
    {'Model': 'Model 1 (Quadratic)', 'Adjusted R2': model_1.rsquared_adj, 'AIC': model_1.aic},
    {'Model': 'Model 2 (Cubic)', 'Adjusted R2': model_2.rsquared_adj, 'AIC': model_2.aic},
    {'Model': 'Model 3 (Logarithmic)', 'Adjusted R2': model_3.rsquared_adj, 'AIC': model_3.aic},
    {'Model': 'Model 4 (Interaction)', 'Adjusted R2': model_4.rsquared_adj, 'AIC': model_4.aic},
    {'Model': 'Model 5 (Quadratic Interaction)', 'Adjusted R2': model_5.rsquared_adj, 'AIC': model_5.aic}
]

# Creating a DataFrame for the results
results_df = pd.DataFrame(results)
results_df
```

## Exercise 5(d) Calculate t-interval

```python
import numpy as np
import scipy.stats as stats

def sim_ci(n, n_reps, dist):
    """
    Simulate confidence intervals for a given distribution.

    Parameters:
    n (int): Sample size for each simulation.
    n_reps (int): Number of repetitions for simulation.
    dist (function): Function to generate random samples.

    Returns:
    np.array: An array of t-intervals.
    """
    ci_array = np.zeros((n_reps, 2))

    for i in range(n_reps):
        sample = dist(n)
        mean = np.mean(sample)
        std_err = stats.sem(sample)
        ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std_err)
        ci_array[i] = ci

    return ci_array

# Example distribution: normal distribution with mean=1 and standard deviation=2
dist = lambda size: np.random.normal(1, 2, size)

# Simulating t-intervals
simulated_cis = sim_ci(10, 3, dist)
simulated_cis
```

## Exercise 5(e) Calculating coverage

```python
import numpy as np
import scipy.stats as stats

def sim_ci(n, n_reps, dist):
    """
    Simulate confidence intervals for a given distribution.
    """
    ci_array = np.zeros((n_reps, 2))
    for i in range(n_reps):
        sample = dist(n)
        mean = np.mean(sample)
        std_err = stats.sem(sample)
        ci = stats.t.interval(0.95, df=n-1, loc=mean, scale=std_err)
        ci_array[i] = ci
    return ci_array

def sim_cov(n, n_reps, dist, mean):
    """
    Simulate coverage of the t-interval.
    """
    ci_array = sim_ci(n, n_reps, dist)
    coverage_array = (ci_array[:, 0] <= mean) & (ci_array[:, 1] >= mean)
    coverage = np.mean(coverage_array)
    return coverage

# Using the specified distribution with seed for reproducibility
dist = lambda size: np.random.default_rng(seed=1).normal(1, 2, size)

# Simulating coverage
simulated_coverage = sim_cov(10, 10000, dist, 1)
simulated_coverage

```

## Exercise 5(f) Simulating multiple condence intervals at once

```python
import numpy as np
import scipy.stats as stats

# Define distributions
distributions = [
    lambda size: np.random.normal(1, 3, size),
    lambda size: np.random.normal(0, 0.1, size),
    lambda size: np.random.exponential(3, size),
    lambda size: np.random.exponential(0.3, size),
    lambda size: np.random.logistic(3, 3, size),
    lambda size: np.random.logistic(0, 0.1, size),
    lambda size: np.random.lognormal(0, 1, size),
    lambda size: np.random.lognormal(2, 2, size)
]

# Calculate means for log-normal distributions
mean_log_normal_0_1 = np.exp(0 + 1**2 / 2)
mean_log_normal_2_2 = np.exp(2 + 2**2 / 2)

# Define means
means = [1, 0, 3, 0.3, 3, 0, mean_log_normal_0_1, mean_log_normal_2_2]

# Sample sizes
sample_sizes = [10, 100]

# Calculate coverages
coverage_results = []
for dist, mean in zip(distributions, means):
    for n in sample_sizes:
        coverage = sim_cov(n, 10000, dist, mean)
        coverage_results.append((dist.__name__, n, coverage))
```
